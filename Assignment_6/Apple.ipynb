```python
og_img=cv2.imread('/content/drive/MyDrive/raptor/shopping.jpeg')
no_of_rotn=4
for i in range(no_of_rotn):
  rot_img=cv2.rotate(og_img,cv2.ROTATE_90_CLOCKWISE)
  cv2.imwrite(f'apple{i*90}.jpeg',rot_img)
  og_img=rot_img.copy()
  cv2_imshow(og_img)
```

```python
height,width=img.shape[:2]
center=(width/2,height/2)

angle=45
rotn_matrix=cv2.getRotationMatrix2D(center,angle,1.0)

rotd_img=cv2.warpAffine(img,rotn_matrix,(width,height))

plt.imshow(img)
plt.imshow(rotd_img)
```

```python
number=10
for i in range(number):
  height,width=img.shape[:2]
  center=(width/2,height/2)

  angle=random.uniform(0,360)
  rotn_matrix=cv2.getRotationMatrix2D(center,angle,1.0)
  rotd_img=cv2.warpAffine(img,rotn_matrix,(width,height))
  cv2.imwrite(f'applesweet{i*1}.jpeg',rotd_img)
  img=rotd_img.copy()

  plt.imshow(img)
  plt.imshow(rotd_img)
```
